# Referenz analyse

Der Ofen macht bereits selber Vorhersagen für den nächsten Nachfüllzeitpunkt. Das neue Modell soll diese Vorhersagen verbessern.
Um dies beurteilen zu können, wird die Vorhersage des neuen Modells mit der Vorhersage des Ofens verglichen. Dafür werden hier zunächst einige Kennzahlen berechnet und visualisiert.

```python
from helper import *

df = load_merged_data(only_turned_off=True, drop_na=True)
```

```python
y_actual = df['refuellingTimePointActual']
y_pred = df['refuellingTimePoint']
```

```python
# plot the data
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

sns.scatterplot(x=y_actual, y=y_pred)
# add reference line with slope of 1
plt.plot([0, 100], [0, 100], color='red', lw=3)
```

```python
from sklearn.metrics import mean_squared_error

mse = mean_squared_error(y_pred, y_actual)
print(f'MSE:  {mse:.2f}')
print(f'RSME: {np.sqrt(mse):.2f}')
```
