# Gegenüberstellung matplotlib und seaborn

Matplotlib wurde hauptsächlich für grundlegende Visualisierung von Daten entwickelt. Die Bibliothek beinhaltet daher grundlegende Diagrammtypen, wie Balken-, Torten-, Linien-, Punkt- bzw. Streudiagramme. Seaborn hingegen bietet eine Vielzahl von Visualisierungsmöglichkeiten bei reduzierter Syntax und optisch ansprechenderen Standardgrafiken. Der Fokus von seaborn liegt dabei auf der Erstellung von statistischen Grafiken. Zusätzlich bietet Seaborn die bessere Integrität mit den Datenrahmen der pandas Bibliothek, die ebenfalls im Bereich der Datenvorverarbeitung dieser Hausarbeit zum Einsatz kommt. In punkto Flexibilität ist matplotlib in hohem Maße anpassbar und leistungsstark und bietet somit im Vergleich zu seaborn einen Vorteil, wenn es um die Individualisierung der Grafiken geht. Je höher jedoch der Individualisierungsgrad der Grafik, desto höher ist ebenso die Komplexität, die bei der Erzeugung der Grafik an den Tag gelegt werden muss. Da seaborn jedoch als Hochschicht API auf matplotlib aufsetzt, kann jede mit seaborn erstellte Grafik ebenso unter Zuhilfenahme der matplotlib Funktionalitäten weiter angepasst werden.

# Was ist seaborn?

Seaborn ist eine Bibliothek zur Erstellung statistischer Grafiken in Python. Sie baut auf matplotlib auf und ist eng mit den Datenstrukturen von pandas integriert.
Seaborn hilft Ihnen beim Erforschen und Verstehen Ihrer Daten. Die Plotting-Funktionen arbeiten mit Dataframes und Arrays, die ganze Datensätze enthalten, und führen intern das notwendige semantische Mapping und die statistische Aggregation durch, um informative Plots zu erstellen. Die datensatzorientierte, deklarative API ermöglicht es Ihnen, sich auf die Bedeutung der verschiedenen Elemente Ihrer Diagramme zu konzentrieren und nicht auf die Details, wie sie zu zeichnen sind.

# Was ist eine explorative Datenanalyse (EDA)?

Die in den 1970er Jahren von dem amerikanischen Mathematiker John Turkey eingeführte explorative Datenanalyse (EDA) ist eine Methodik zur umfassenden Untersuchung von Datensätzen, die auch heute noch allgegenwertig ist. Hauptaugenmerk liegt hierbei auf der Herausstellung und Zusammenfassung grundlegender Eigenschaften des Datensatzes. Dazu zählen Trends, Muster und Beziehungen, die nicht ohne weiteres erkennbar sind. Damit diese Eigenschaften leichter zu erkennen sind, wird häufig auf Datenvisualisierungstechniken zurückgegriffen. 

Die EDA führt zu einem besseren Verständnis der Variablen, aus denen der Datensatz besteht, und verdeutlicht die Zusammenhänge zwischen ihnen. Darüber hinaus kann die EDA Aufschluss darüber geben, welche statistischen Methoden für die weitergehende Datenanalyse geeignet sein können. 

[Explorative Datenanalyse](https://www.ibm.com/de-de/topics/exploratory-data-analysis)

# Warum ist eine EDA wichtig in der Datenwissenschaft?

Es soll zunächst ein erstes Verständnis für die Merkmale eines Datensatzes geschaffen werden. Die EDA gibt einen Überblick über die Verteilung, die zentralen Tendenzen und die Schwankungen der Daten und bietet Einblicke in die Natur der vorliegenden Informationen. Zusätzlich können Datenwissenschaftler mittels einer EDA Muster, Trends und Unregelmäßigkeiten in den Daten erkennen. Dies hilft bei der Erkennung von Anomalien oder Ausreißern, die beispielsweise auf Probleme mit der Datenqualität hinweisen oder einzigartige Einblicke bieten können. Anhand einer EDA lassen sich zudem auch leichter Hypothesen über die Beziehungen und Trends in den Daten aufstellen, welche in der anschließenden statistischen Analyse formell getestet werden können.

# Welche Arten der EDA gibt es? 

Grundlegend kann eine explorative Datenanalyse auf 2 Arten durchgeführt werden. Quantitativ durch beschreibende Statistik oder durch grafische Aufbereitung der Daten. In dieser Arbeit steht die visuelle Aufbereitung der Daten im Fokus und wird im Folgenden näher erläutert. Bei der grafischen EDA wird zwischen der univariaten Analyse und der multivariaten Analyse unterschieden. Die univariate Analyse fokussiert lediglich eine Variable des Datensatzes zur Zeit. Für kategorische Variablen bieten sich daher Balkendiagramme mit der absoluten Häufigkeit der einzelnen Kategorien an, oder Tortendiagramme, die zusätzliche eine prozentuale Verteilung der einzelnen Kategorien der zu untersuchenden Variable darstellen können. Handelt es sich hingegen um eine numerische Variable, die untersucht werden soll so kann beispielsweise ein Histogramm Aufschluss über die Verteilung der Daten gegeben. Einen ähnlichen Einblick bietet, basierend auf einem Histogramm, eine zusätzliche Dichteverteilung in Form einer Kurve über dem Histogramm. Eine andere Art, die Verteilung innerhalb der Daten zu visualisieren, bietet ein Boxplot Diagramm. Es gibt bspw. Aufschluss über Minimal- und Maximalwerte, Quantile, den Mittelwert bzw. Median der Daten und es unterstützt visuell bei der Erkennung von möglichen Ausreißen in den Daten. Ähnlich zum Boxplot Diagramm, kann auch ein Violinen Diagramm zur Darstellung der Verteilung der Daten verwendet werden.   
Im Gegensatz zur univariaten Analyse befasst sich die multivariate Analyse mit mindestens 2 Variablen zur selben Zeit. Es kann nun also bspw. die Korrelation der Variablen untersucht werden. Auch hier gibt es wieder unterschiedliche Darstellungsmöglichkeiten, je nachdem, ob es sich um kategorische oder numerische Variablen handelt. Hinzu kommt, dass auch Kombinationen aus kategorischen und numerischen Variablen untersucht werden können. 
Für die weiteren Erläuterungen wird sich auf die bivariate Analyse als Spezialfall der multivariaten Analyse fokussiert, wo lediglich 2 Variablen zu gleichen Zeit untersucht werden. Es ergeben sich somit die folgenden 3 Kombinationen, die berücksichtigt werden müssen. 
Zunächst sollen zwei numerische Variablen miteinander verglichen werden. Eines der gängigsten Diagramme für diesen Anwendungsfall ist das Streudiagramm, welches den Schnittpunkt zwischen den beiden Variablen als Punkt in einem 2-dimensionalen Koordinatensystem mit den beiden Variablen als X- bzw. als Y-Achse abbildet. 
Soll die Beziehung zwischen einer numerischen und einer kategorischen Variable dargestellt werden, so bietet sich ein klassisches Balkendiagramm an. Jeder Balken repräsentiert eine Ausprägung der kategorischen Variable und die Länge bzw. Höhe des Balkens repräsentiert den zugehörigen Wert der numerischen Variable zu dieser Kategorie. 
Abschließend wird noch die Beziehung zwischen zwei kategorischen Variablen beschrieben. In diesem Fall ist eine gängige Visualisierungsmöglichkeit, die Gegenüberstellung der beiden Variablen in einer Heatmap. Hier wird die Korrelation der beiden Variablen anhand einer Farbskala dargestellt.
